**RESTAURANT MENU & ORDER MANAGEMENT SYSTEM**

Task 1: Menu Exploration

```python
import copy
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

categories = ["Starters", "Mains", "Desserts"]
for cat in categories:
    print(f"\n===== {cat} =====")
    for item, details in menu.items():
        if details["category"] == cat:
            status = "Available" if details["available"] else "Unavailable"
            print(f"{item:<15} \u20b9{details['price']:.2f}   [{status}]")
```

```python
# ------------------ Menu Statistics ------------------

# Print total number of items in the menu
print("Total items:", len(menu))

# Count how many items are available using generator expression
available_count = sum(1 for item in menu.values() if item["available"])
print("Available items:", available_count)

# ------------------ Most Expensive Item ------------------

# Find the item with the highest price using max() and lambda
exp_item = max(menu.items(), key=lambda x: x[1]["price"])

# Display most expensive item with its price
print(f"Most expensive: {exp_item[0]} — \u20b9{exp_item[1]['price']:.2f}")

# ------------------ Items Under ₹150 ------------------

# Display items whose price is less than ₹150
print("\nItems under \u20b9150:")
for item, details in menu.items():

  # Check price condition
    if details["price"] < 150:
        print(f"  {item:<15} \u20b9{details['price']:.2f}")
```

Task 2: Cart Operation

```python
cart = []

def add_item(item_name, qty):
    if item_name not in menu:
        print(f"  \u2717 '{item_name}' not found in menu")
        return
    if not menu[item_name]["available"]:
        print(f"  \u2717 '{item_name}' is currently unavailable")
        return
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += qty
            print(f"  \u2713 Updated '{item_name}' quantity to {item['quantity']}")
            return
    cart.append({"item": item_name, "quantity": qty, "price": menu[item_name]["price"]})
    print(f"  \u2713 Added '{item_name}' x{qty}")

def remove_item(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            print(f"  \u2713 Removed '{item_name}'")
            return
    print(f"  \u2717 '{item_name}' not in cart")

def show_cart():
    print("  Cart:", cart if cart else "(empty)")
```

```python
# Simulation
print("--- Cart Simulation ---")

print("\nStep 1 — Add Paneer Tikka x2")
add_item("Paneer Tikka", 2)
show_cart()

print("\nStep 2 — Add Gulab Jamun x1")
add_item("Gulab Jamun", 1)
show_cart()

print("\nStep 3 — Add Paneer Tikka x1 (should update to 3)")
add_item("Paneer Tikka", 1)
show_cart()

print("\nStep 4 — Try Mystery Burger (doesn't exist)")
add_item("Mystery Burger", 1)
show_cart()

print("\nStep 5 — Try Chicken Wings (unavailable)")
add_item("Chicken Wings", 1)
show_cart()

print("\nStep 6 — Remove Gulab Jamun")
remove_item("Gulab Jamun")
show_cart()
```

```python
# Order Summary
print("========== Order Summary ==========")
subtotal = 0
for item in cart:
    total_price = item["quantity"] * item["price"]
    subtotal += total_price
    print(f"{item['item']:<20} x{item['quantity']}   \u20b9{total_price:.2f}")
print("------------------------------------")
gst   = round(subtotal * 0.05, 2)
total = round(subtotal + gst, 2)
print(f"{'Subtotal:':<28} \u20b9{subtotal:.2f}")
print(f"{'GST (5%):':<28} \u20b9{gst:.2f}")
print(f"{'Total Payable:':<28} \u20b9{total:.2f}")
print("====================================")
```

Task 3: Inventory Tracker with Deep Copy

```python
import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

inventory_backup = copy.deepcopy(inventory)

# Demonstrate deep copy
print("--- Deep Copy Demo ---")
inventory["Paneer Tikka"]["stock"] = 5
print(f"  inventory['Paneer Tikka']['stock']        = {inventory['Paneer Tikka']['stock']}  (modified)")
print(f"  inventory_backup['Paneer Tikka']['stock'] = {inventory_backup['Paneer Tikka']['stock']}  (backup unaffected)")

# Restore
inventory = copy.deepcopy(inventory_backup)
print("  \u2713 inventory restored to original.")
```

```python
# Deduct cart stock
print("Deducting cart quantities...")
for item in cart:
    name = item["item"]
    qty  = item["quantity"]
    if inventory[name]["stock"] >= qty:
        inventory[name]["stock"] -= qty
        print(f"  \u2713 Deducted {qty} from '{name}'. Remaining: {inventory[name]['stock']}")
    else:
        print(f"  \u26a0 Not enough stock for '{name}'. Deducting all available ({inventory[name]['stock']}).")
        inventory[name]["stock"] = 0

# Reorder alerts
print("\nReorder Alerts:")
alert_found = False
for item, data in inventory.items():
    if data["stock"] <= data["reorder_level"]:
        print(f"  \u26a0 Reorder Alert: {item} — Only {data['stock']} unit(s) left (reorder level: {data['reorder_level']})")
        alert_found = True
if not alert_found:
    print("  No items need reordering.")

# Confirm backup differs
print(f"\n  inventory['Paneer Tikka']['stock']        = {inventory['Paneer Tikka']['stock']}")
print(f"  inventory_backup['Paneer Tikka']['stock'] = {inventory_backup['Paneer Tikka']['stock']}")
print("  \u2713 Deep copy confirmed — backup is unaffected.")
```

Task 4: Daily Sales Log Analysis

```python
sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

def print_revenue_table(log):
    best_day   = None
    best_total = 0
    print(f"  {'Date':<14} {'Revenue':>10}")
    print("  " + "-" * 26)
    for date in sorted(log):
        day_rev = sum(order["total"] for order in log[date])
        print(f"  {date:<14} \u20b9{day_rev:>9.2f}")
        if day_rev > best_total:
            best_total = day_rev
            best_day   = date
    print(f"\n  Best-selling day: {best_day} — \u20b9{best_total:.2f}")

print("--- Revenue Per Day (original) ---")
print_revenue_table(sales_log)
```

```python
# Most ordered item
item_count = {}
for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1
most_ordered = max(item_count, key=item_count.get)
print(f"Most ordered item: {most_ordered} ({item_count[most_ordered]} orders)")
```

```python
# Add new day and reprint
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("--- Revenue Per Day (after adding 2025-01-05) ---")
print_revenue_table(sales_log)
```

```python
# Enumerate all orders
print("All Orders:")
all_orders = []
for date in sorted(sales_log):
    for order in sales_log[date]:
        all_orders.append((date, order))

for idx, (date, order) in enumerate(all_orders, start=1):
    items_str = ", ".join(order["items"])
    print(f"  {idx}. [{date}] Order #{order['order_id']} — \u20b9{order['total']:.2f} — Items: {items_str}")
```
